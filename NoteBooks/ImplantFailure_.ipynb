```python
def format_describe_table(df):
    formatted_table = df.describe().T

    # Add additional statistics you want to display, if any
    formatted_table['median'] = df.median()
    formatted_table['skewness'] = df.skew()
    formatted_table['kurtosis'] = df.kurtosis()
    formatted_table['missing_count'] = df.isnull().sum()
    formatted_table['unique_values'] = df.nunique()

    # Customize the column names for better readability
    formatted_table.rename(columns={
        'count': 'Count',
        'mean': 'Mean',
        'std': 'Standard Deviation',
        'min': 'Minimum',
        '25%': '25th Percentile',
        '50%': 'Median',
        '75%': '75th Percentile',
        'max': 'Maximum',
        'skewness': 'Skewness',
        'kurtosis': 'Kurtosis',
        'missing_count': 'Missing Count',
        'unique_values': 'Unique Values'
    }, inplace=True)

    # Format numbers to desired precision and alignment
    formatted_table = formatted_table.applymap(lambda x: f'{x:.2f}' if isinstance(x, (int, float)) else x)

    return formatted_table
```

# Test for Normality

```python
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt


numerical_columns = ['Age', 'Gender', 'Systemic disease', 'Factors of missing', 'Tobacco smoking',
                     'Betel nut Chewing', 'Alcohol consumption', 'Departments', 'Surgeon experience',
                     'Location of implant', 'Bone density', 'Timing of implant placement', 'Ridge augmentation',
                     'Maxillary sinus augmentation', 'Implant system', 'Types of prosthesis',
                     'Angle of abutment', 'Prosthesis fixation', 'Dental implant failure']

for column in numerical_columns:
    # Shapiro-Wilk Test
    stat, p = stats.shapiro(df[column])
    print(f"Shapiro-Wilk Test for {column}:")
    print(f"  Statistic: {stat}")
    print(f"  p-value: {p}")
    if p > 0.05:
        print("  Result: Data is normally distributed (fail to reject null hypothesis)")
    else:
        print("  Result: Data is not normally distributed (reject null hypothesis)")

    # QQ plot
    plt.figure(figsize=(8, 6))
    stats.probplot(df[column], dist="norm", plot=plt)
    plt.title(f'Q-Q Plot of {column}')
    plt.show()

    # Normal distribution graph
    plt.figure(figsize=(8, 6))
    sns.histplot(df[column], kde=True, color='blue')
    plt.title(f'Normal Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Density')
    plt.show()
```

# Independent T- Test

```python
from scipy.stats import ttest_ind

# Separate the data for the two groups (Male and Female)
group1 = df[df['Gender'] == 1]['Age']
group2 = df[df['Gender'] == 0]['Age']

# Perform the Independent t-test
t_stat, p_value = ttest_ind(group1, group2)
alpha = 0.05

# Print the results
print("Independent t-test:")
print()
print("Null Hypothesis (H0): There is no difference in the means of Age between Male and Female.")
print("Alternative Hypothesis (H1): There is a significant difference in the means of Age between Male and Female.")
print()
print("T-statistic:", t_stat)
print("P-value:", p_value)
print()

# Check if the p-value is less than the significance level
if p_value < alpha:
    print("Result: Reject the Null Hypothesis (H0). There is a significant difference in the means of Age between Male and Female.")
else:
    print("Result: Fail to reject the Null Hypothesis (H0). There is no significant difference in the means of Age between Male and Female.")
```

# Chi-square Test:

```python
# Function to encode age groups
def encode_age_group(age):
    if age >= 31 and age <= 50:
        return 'Middle Aged Adults'
    else:
        return 'Old Adults'

# Apply the encoding function to create a new column 'age_group'
df['age_group'] = df['Age'].apply(encode_age_group)

# Create a new DataFrame with only 'age' and 'age_group' columns
df_updated = df[['Age', 'age_group', 'Bone density','Location of implant','Factors of missing','Systemic disease']]

# Display the updated DataFrame
df_updated.head()
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

#Chi-square to check upon the association between Gender and Bone Density

#H0: There is no association between 'Age Group' and 'Bone density' concerning dental implant failure.
#H1: There is association between 'Age Group' and 'Bone density' concerning dental implant failure.

#H0: There is no association between 'Age Group' and 'Systemic disease' concerning dental implant failure.
#H1: There is association between 'Age Group' and 'Systemic disease' concerning dental implant failure.

#H0: There is no association between 'Age Group' and 'Factors of missing' concerning dental implant failure.
#H1: There is association between 'Age Group' and 'Factors of missing' concerning dental implant failure.

Alpha = 0.05



attribute_combinations = [('age_group', 'Bone density'),
                          ('age_group', 'Systemic disease'),
                          ('age_group', 'Factors of missing')]


for attribute1, attribute2 in attribute_combinations:
    contingency_table = pd.crosstab(df_updated[attribute1], df_updated[attribute2])
    chi2, p_value, _, _ = chi2_contingency(contingency_table)

    print(f"Chi-square test between {attribute1} and {attribute2}:")
    print(f"Chi-square statistic: {chi2}")
    print(f"P-value: {p_value}")
    print("Result: Significant" if p_value > 0.01 else "Result: Not Significant")
    print()
```

**Age Group and Systemic Disease:**

Therefore, we reject the null hypothesis (H0) and conclude that there is a significant association between 'Age Group' and 'Systemic Disease' concerning dental implant failure.

Therefore, we reject the null hypothesis (H0) and conclude that there is a significant association between 'Age Group' and 'Factors of Missing' concerning dental implant failure
